# **Breast Cancer Prediction using ML**

```python
import numpy as np
import pandas as pd
```

```python
breast = pd.read_csv('breast cancer.csv')
breast.head()
```

```python
breast['diagnosis'].value_counts()
```

```python
breast.shape
```

```python
breast.isnull().sum()
```

```python
breast.duplicated().sum()
```

```python
breast.info()
```

```python
# Drop non-numeric columns 'id' and 'diagnosis' before calculating correlation
breast_numeric = breast.drop(['diagnosis'], axis=1)
breast_numeric.corr()
```

```python
breast.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
breast.describe()
```

```python
#Encoding the target Variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
breast['diagnosis'] = breast['diagnosis'].map({'M':1, 'B':0})
```

```python
breast['diagnosis'].value_counts()
```

```python
breast['diagnosis'] = le.fit_transform(breast['diagnosis'])
```

```python
x = breast.drop('diagnosis', axis=1)
y = breast['diagnosis']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

# **Training Model**

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(x_train, y_train)
y_pred = lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
